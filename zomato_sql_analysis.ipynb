```python
import sqlite3

# this creates a database file called Zomato_project.db in your current folder
```

```python
conn = sqlite3.connect("zomato_project.db")
```

```python
cursor = conn.cursor()

print("SQLite database created and connected!")
```

```python
# Use your loaded CSV (df)
cols = df.columns
# Make all columns TEXT type for simplicity
col_str = ", ".join([f"{col} TEXT" for col in cols])
table_name = "restaurants"

# Remove old table if it exists
cursor.execute(f"DROP TABLE IF EXISTS {table_name}")

# Create new table
cursor.execute(f"CREATE TABLE {table_name} ({col_str})")
print(f"Table '{table_name}' created successfully!")
```

```python
# Insert all rows from the CSV into the SQLite table
for i, row in df.iterrows():
    sql = f"INSERT INTO {table_name} ({', '.join(cols)}) VALUES ({', '.join(['?']*len(cols))})"
    cursor.execute(sql, tuple(row))

# Save changes
conn.commit()
print("CSV imported into SQLite successfully!")
```

```python
# Read first 5 rows from SQL
df_sql = pd.read_sql(f"SELECT * FROM {table_name} LIMIT 5;", conn)
df_sql
```
