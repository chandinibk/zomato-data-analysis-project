### DATA CLEANING

```python

print("missing values")
print(df.isnull().sum())
```

```python
df['phone']= df['phone'].fillna("Not Available")
```

```python
df['location'] = df['location'].fillna(df['location'].mode()[0])
```

```python
df['rest_type'] = df['rest_type'].fillna(df['rest_type'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['dish_liked'] = df['dish_liked'].fillna('not mentioned')
```

```python
df.isnull().sum()
```

```python
df['cuisines'] = df['cuisines'].fillna(df['cuisines'].mode()[0])
```

```python
# cleaned 'approx_cost(for two people)' by removing commas, converting to numeric,and handling missing values

df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',','')
```

```python
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'],errors = 'coerce')
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].median())
```

### FEATURE ENGINEERING

```python
# 1.COST CATEGORY

df['cost_category'] = pd.cut(df['approx_cost(for two people)'],bins=[0, 300, 600, 1000, 5000],labels=['Budget', 'Affordable', 'Expensive', 'Luxury'])
```

```python
# 2.ONLINE ORDER FLAG(BINARY)

df['online_order_flag'] = df['online_order'].map({'Yes': 1, 'No': 0})
```

```python
# 3.TABLE BOOKING FLAG(BINARY)

df['table_booking_flag'] = df['book_table'].map({'Yes': 1, 'No': 0})
```

```python
# 4.VOTES LOG TRANSFORMATION

import numpy as np
df['votes_log'] = np.log1p(df['votes'])
```

### VISUALIZATION

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(x='cost_category', data=df, palette='viridis', order=df['cost_category'].value_counts().index)
plt.title('Number of Restaurants by Cost Category', fontsize=14)
plt.xlabel('Cost Category')
plt.ylabel('Count')
plt.show()
```

insights

1.Affordable and Budget restaurants dominate the dataset.
2.Expensive and Luxury restaurants form a much smaller share.
3.Indicate Zomato primarily caters to price-sensitive, mass customers.

```python
df['online_order_flag'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue','lightpink'], startangle=90)
plt.title('Restaurants Offering Online Orders', fontsize=14)
plt.ylabel('')
plt.show()
```

Insights.

1.~59% of restaurants offer online ordering, while ~41% do not.
2.Online delivery is common but not yet universal.
3.There is clear scope to onboard more restuarants to online delivery.

```python
df['table_booking_flag'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen','pink'], startangle=90, explode=(0.05,0))
plt.title('Restaurants Offering Table Booking', fontsize=14)
plt.ylabel('')
plt.show()
```

Insights.

1.Only ~12.5% of restaurants offer table booking.
2.Majority of restaurants focus on walk-in or delivery-based models.
3.Table booking is likely assocaited with premium or dine-in focused restaurants.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['votes_log'], bins=30, kde=True, color='purple')
plt.title('Distribution of Votes (Log Transformed)', fontsize=14)
plt.xlabel('Log of Votes')
plt.ylabel('Frequency')
plt.show()
```

Summary.

Zomato's platform is dominated by affordable restaurants, with moderate adoption of online delivery and limited availability of online delivery
and limited availability of table booking, indicating strong focus on mass food delivery rather than premium dine-in experiences.


Insights.

1.Votes are highly skewed, with most restaurants recieving low votes.
2.Only a small fraction of restaurants attract very high customers engagement.
3.Log transformation reveals long-tail popularity in restaurant voting.
